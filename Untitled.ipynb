# In This we are visulaizing Swiggy dataset 

```python
#importing libraries for visulization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns    
```

```python
plt.style.use("ggplot")            
sns.set_context("talk")            # readable font sizes
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
customers   = pd.read_csv("swiggy dataset for analysis/Customers Data.csv")
partners    = pd.read_csv("swiggy dataset for analysis/Delivery Partners Data.csv")
menu        = pd.read_csv("swiggy dataset for analysis/Menu Items Data.csv")
orders      = pd.read_csv("swiggy dataset for analysis/Orders Data.csv")
restaurants = pd.read_csv("swiggy dataset for analysis/Restaurants Data.csv")
reviews     = pd.read_csv("swiggy dataset for analysis/Reviews and Ratings Data.csv")
```

```python
orders.columns = (
    orders.columns.str.strip()          # remove leading/trailing spaces
                   .str.lower()         
                   .str.replace(" ", "_", regex=False)
)
```

```python
orders["order_time"] = pd.to_datetime(
    orders["order_date"] + " " + orders["order_time"],
    format="%d-%m-%Y %I:%M %p",
    errors="coerce"
)
```

```python
orders["order_total"] = (
    orders["order_total"]
      .str.replace(r"[$,]", "", regex=True)     # remove $ and commas
      .astype(float)
)
```

```python
#view the data snapshots
print("\n Orders snapshot:")
print(orders.head(3), "\n")

print("Missing values (%):")
print((orders.isnull().mean()*100).round(2))
```

```python
# Counting Daily Order
daily_orders = (
    orders.set_index("order_time")
          .resample("D")["order_id"]
          .count()
          .rename("num_orders")
)
```

```python
# Revenue Calculation 
revenue_by_restaurant = (
    orders.groupby("restaurant_id")["order_total"]
          .sum()
          .sort_values(ascending=False)
)
```

```python
# Daily Order Trend
daily_orders.plot()
plt.title("Daily Order Volume")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.tight_layout()
plt.show()
```

```python
# Top 15 Restaurants by Revenue
top15 = revenue_by_restaurant.head(15)
sns.barplot(x=top15.values, y=top15.index, palette="crest")
plt.title("Top 15 Restaurants by Total Revenue")
plt.xlabel("Revenue ()")
plt.ylabel("Restaurant ID")
plt.tight_layout()
plt.show()
```

```python
# Order-Value Distribution
sns.histplot(orders["order_total"], bins=40, kde=True, color="steelblue")
plt.title("Distribution of Order Values")
plt.xlabel("Order Total ()")
plt.tight_layout()
plt.show()
```

```python
# Payment-Method Breakdown
sns.countplot(
    data=orders,
    y="payment_method",
    order=orders["payment_method"].value_counts().index,
    palette="pastel"
)
plt.title("Payment Method Usage")
plt.xlabel("Number of Orders")
plt.ylabel("Payment Method")
plt.tight_layout()
plt.show()
```

```python
# 6-5   Order-Status Breakdown
sns.countplot(
    data=orders,
    x="order_status",
    order=orders["order_status"].value_counts().index,
    palette="Set2"
)
plt.title("Order Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
#  Hour-of-Day Heatmap
orders["hour"] = orders["order_time"].dt.hour
hourly = (
    orders.groupby("hour")["order_id"]
          .count()
          .reindex(range(0, 24), fill_value=0)
)
```

```python
sns.heatmap(hourly.to_frame().T, cmap="rocket", annot=True, fmt="d")
plt.title("Orders per Hour of Day")
plt.ylabel("")            # single-row heatmap; y-axis label redundant
plt.xlabel("Hour (24h)")
plt.tight_layout()
plt.show()
```
